```python
df_hist = df_cocktail.loc[df_cocktail.index <= hist_cutoff].copy()
df_synth = df_cocktail.loc[df_cocktail.index > hist_cutoff].copy()

fig = go.Figure()

colors = {
    "HH": "#0C1BF2",
    "ttf_usd_mmbtu": "#39F1FA",
    "JKM": "#FF6F00",
    "Coal FOB Australia (USD/MMBtu)": "#5B8C5A",
    "Coal CIF (USD/MMBtu)": "#886338",
    "Oil CIF (USD/MMBtu)": "#B85C38",
    "LNG CIF (USD/MMBtu)": "#1F7A8C",
    "coal_aus_fob_usd_t": "#5B8C5A",
    "coal_zad_fob_usd_t": "#8FB339",
    "coal_jpn_cif_usd_t": "#886338",
    # "coal_jpn_cif_jpy_t": "#6B4F4F",
    "coal_ind_fob_usd_t": "#C97B84",
    # "oil_jpn_cif_jpy_kl": "#7D5BA6",
    "oil_jpn_cif_usd_kl": "#B85C38",
    # "lng_jpn_cif_jpy_t": "#0B6E4F",
    # "lng_jpn_cif_usd_t": "#2A9D8F",
    # "lng_jpn_cif_usd_mmbtu": "#1F7A8C",
}

legendonly_cols = {
    "coal_aus_fob_usd_t",
    "coal_zad_fob_usd_t",
    "coal_jpn_cif_usd_t",
    # "coal_jpn_cif_jpy_t",
    "coal_ind_fob_usd_t",
    # "oil_jpn_cif_jpy_kl",
    "oil_jpn_cif_usd_kl",
    # "lng_jpn_cif_jpy_t",
    # "lng_jpn_cif_usd_t",
    # "lng_jpn_cif_usd_mmbtu",
}

for col in plot_df.columns:
    if col not in colors:
        continue
    fig.add_trace(go.Scatter(
        x=plot_df.index,
        y=plot_df[col],
        name=col,
        mode="lines",
        visible="legendonly" if col in legendonly_cols else True,
        line=dict(color=colors[col], width=1.8),
    ))

fig.add_trace(go.Scatter(
    x=df_hist.index,
    y=df_hist["jcc_usd_mmbtu"],
    name="JCC Hist",
    mode="lines",
    line=dict(color="#B22222", width=2.2),
))

# if not df_synth.empty:
#     fig.add_trace(go.Scatter(
#         x=df_synth.index,
#         y=df_synth["jcc_usd_mmbtu"],
#         name="JCC Synth",
#         mode="markers",
#         line=dict(color="#B22222", width=1.5),
#     ))

fig.add_trace(go.Scatter(
    x=df_hist.index,
    y=df_hist["jlc_usd_mmbtu"],
    name="JLC Hist",
    mode="lines",
    line=dict(color="#4B878E", width=2.2),
))
# vertical line on the 1st of the last month
fig.add_shape(
    type="line",
    x0=(plot_df.index[-1].replace(day=1) - pd.DateOffset(months=1)),
    y0=0,
    x1=(plot_df.index[-1].replace(day=1) - pd.DateOffset(months=1)),
    y1=80,
    line=dict(color="black", width=3, dash="dash"),
)

# if not df_synth.empty:
#     fig.add_trace(go.Scatter(
#         x=df_synth.index,
#         y=df_synth["jlc_usd_mmbtu"],
#         name="JLC Synth",
#         mode="markers",
#         line=dict(color="#4B878E", width=1.5),
#     ))

fig.update_layout(
    title="Fuel prices - historical/synthetic with coal, oil and LNG daily inputs",
    yaxis_title="EUR / MWh or USD / MMBtu",
    template="plotly_white",
    legend=dict(orientation="h", y=-0.18),
    height=560,
    width=980,
    xaxis=dict(range=[pd.Timestamp("2019-10-01"), df_cocktail.index.max()]),
    yaxis=dict(range=[0, 80]),
)
fig.show()
```
